### Analisis de datos inicial

```python
#Identificar los tipos de estados del clima que ofrece el DF en este caso Bad se asume como 
print(df['weather_conditions'].unique())
```

```python
# Convertir la columna de fecha y hora a tipo datetime
df['start_ts'] = pd.to_datetime(df['start_ts'])

# Filtrar los datos para sábados con condiciones climáticas malas y otros días
df['day_of_week'] = df['start_ts'].dt.day_name()
df['is_bad_weather'] = df['weather_conditions'] == 'Bad'

# Filtrar para sábados con mal tiempo
saturdays_bad_weather = df[(df['day_of_week'] == 'Saturday') & (df['is_bad_weather'])]
# Filtrar para otros días
other_days = df[~((df['day_of_week'] == 'Saturday') & (df['is_bad_weather']))]

# Verificar la cantidad de datos
print(f"Número de viajes en sábados con mal tiempo: {len(saturdays_bad_weather)}")
print(f"Número de viajes en otros días: {len(other_days)}")
```

```python
# Eliminar NaN en la duración de los viajes
saturdays_bad_weather = saturdays_bad_weather.dropna(subset=['duration_seconds'])
other_days = other_days.dropna(subset=['duration_seconds'])

# Verificar nuevamente la cantidad de datos después de eliminar NaN
print(f"Número de viajes en sábados con mal tiempo después de limpiar: {len(saturdays_bad_weather)}")
print(f"Número de viajes en otros días después de limpiar: {len(other_days)}")

# Prueba de Levene
if len(saturdays_bad_weather) > 1 and len(other_days) > 1:  # Asegúrate de que haya suficientes datos
    levene_stat, levene_p = stats.levene(saturdays_bad_weather['duration_seconds'], other_days['duration_seconds'])

    # Realizar la prueba t
    t_stat, p_value = stats.ttest_ind(saturdays_bad_weather['duration_seconds'], other_days['duration_seconds'], equal_var=levene_p > 0.05)

    # Resultados
    print(f"Prueba de Levene: estadístico = {levene_stat}, p-valor = {levene_p}")
    print(f"Prueba t: estadístico t = {t_stat}, p-valor = {p_value}")

    # Interpretar los resultados
    alpha = 0.05  # Nivel de significación
    if p_value < alpha:
        print("Rechazamos la hipótesis nula: hay una diferencia significativa en la duración de los viajes.")
    else:
        print("No rechazamos la hipótesis nula: no hay evidencia suficiente para afirmar que hay una diferencia significativa.")
else:
    print("No hay suficientes datos para realizar la prueba.")
```
